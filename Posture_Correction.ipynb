### Calculating Angle Function

We'll be calculating the angle between 3 landmarks. The problem can be considered as finding the angle between two vectors.<br>

The formula used to calculate this angle is given below:

<img src="https://www.onlinemath4all.com/images/anglebetweentwovectors.png" width="200" height="200">


```python
def calculateAngle(landmark1, landmark2, landmark3):  #Function to calculate angle between 3 landmarks 
   
    arr1 = np.asarray(landmark1)     #Considering only the x and y coordinates
    arr1 = arr1[0:2]
    
    arr2 = np.asarray(landmark2)
    arr2 = arr2[0:2]
    
    arr3 = np.asarray(landmark3)
    arr3 = arr3[0:2]
    
    ba = arr1 - arr2
    bc = arr3 - arr2
    
    cosine_angle = np.dot(ba, bc) / (np.linalg.norm(ba) * np.linalg.norm(bc))  #Applying the mathematical formula 
    angle = np.arccos(cosine_angle)
    
    angle = np.degrees(angle)  #Convert it into degrees 
    angle = math.trunc(angle)  #Truncate the result
    
    return(angle)   #Return the angle 
```

Not all of these angles are relevant to the task at hand, which is recognizing the frames in which the sitting posture is incorrect. <br> We need to be careful while picking our classfiying criteria.
<br> 

For the purpose of the task, I have considered 2 angles: the first one between the shoulder, hip and knee and the second one between the hip, knee and ankle. <br> 

As mentioned in the video the correct sitting posture involves the back being straight and the knee making a right angle.
(https://www.youtube.com/watch?v=OyK0oE5rwFY)


We will thus consider that if the angle between the shoulder, hip and knee becomes greater than 105 degrees or less than 90 degrees or the angle between the hip, knee and ankle does not lie in the range of 90 (plus minus 5 degrees) degrees, the posture is a bad one
<br> <br>
